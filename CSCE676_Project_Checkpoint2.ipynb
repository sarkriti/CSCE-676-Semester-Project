# Project Scope and EDA Summary

**Dataset Description**

For this project we will be analyzing infectious disease reportings for the state of California from the years 2001 - 2023. Variables in this dataset include Disease, County, Year, Sex, Cases, Population, Rate, Incidence Rate, Lower_95_CLI, and Upper_95_CLI. Additionally we have feature engineered a new variable called transmission category that classifies the transmission method for each disease (foodborne, waterborne, seawater exposure). There are 199,125 entries in the dataset.

# Research Question 1

## Question: Are counties that experience high infection rates of Coccidioidomycosis (Valley Fever) among men counties where the primary industry of the population are outdoor male dominated fields (construction, agriculture, etc.)?

**Data Mining Task Types:** Anomaly Detection

**Relevant Algorithms:** Isolation Forest (course),Extreme Value Analysis/Mahalanobious Distance (course), One-Class SVM (external)

**Evaluation Criteria:** Anomaly Score, Threshold, Contamination Rate, Z-Scores, Covariance




## Motivation and Feasibility

Motivation: EDA shows that there is commonly a gender disparity among Valley Fever infections where men are more likely to be infected than women. Additionally, the counties with the highest cases are not the most populous counties.This is significant due to the fact that the infection rate for a lot of common diseases correspond to counties with larger populations. The counties with high valley fever infection rates suggest that this may be due to unique environmental and industrial factors.

Non-triviality: By identifying areas where a specific population subset is more at risk for a disease (men in agriculture, men in construction, etc.), we can use this information to develop more informed health policies to combat potential epidemic issues.

Feasibility: Data mining techinques are easy to learn and implement

Risks: Information regarding county industries are not available in the current data set and will need to feature engineered/linked with another dataset. Parameter sensitivity, false positive rate.

## Methodlogical Planning

Course Algorithms: Isolation Forest, Extreme Value Analysis, Mahalabonious Distance

External Algorithms: One-Class SVM

Evaluation: Anamoly Scores, Contamination Rate, Derived Patterns

Baseline: Male:Female ratio of infection rates of Valley Fever for each county
